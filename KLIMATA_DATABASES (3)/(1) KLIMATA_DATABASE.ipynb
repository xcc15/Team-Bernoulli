## KLIMATA ILOILO DATABASE CREATION

###  After the data preprocessing, we created a SQL database for the datasets to store, organize, and efficiently manage multiple related tables so they can be easily queried, linked, and updated during analysis.

```python
import os
import sqlite3
import pandas as pd

# Folder where your CSV files are stored
folder_path = r"C:\Users\Value Lines\Documents\FINAL_DATASETS"

#  Create or connect to a local SQLite database
conn = sqlite3.connect("final_datasets.db")
cursor = conn.cursor()

# Loop through all CSV files in the folder
for file in os.listdir(folder_path):
    if file.endswith(".csv"):
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)

        # Use file name (without extension) as table name
        table_name = os.path.splitext(file)[0]

        # Save DataFrame as a SQL table
        df.to_sql(table_name, conn, if_exists="replace", index=False)
        print(f" Saved {table_name} to database")

conn = sqlite3.connect("final_datasets.db")
cursor = conn.cursor()

# Confirm saved tables
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()
print("\nðŸ“‹ Tables in database:")
for t in tables:
    print("-", t[0])

# Close the connection
conn.close()
print("\n Database created successfully: final_datasets.db")
```

```python
# Check if tables is a DataFrame or list

conn = sqlite3.connect("final_datasets.db")

if isinstance(tables, pd.DataFrame):
    table_names = tables['name'].tolist()
else:
    table_names = [t[0] for t in tables]  # if it's a list of tuples

# Loop through and preview first 5 rows of each table
for table in table_names:
    print(f"\nðŸ“„ First 5 rows of table: {table}")
    preview = pd.read_sql(f"SELECT * FROM {table} LIMIT 5;", conn)
    print(preview)
    
conn.close()
```
