## KLIMATA ILOILO DATA PREPROCESSING (STANDARDIZATION STAGE)

### For the next preprocessing step, we standardized the datasets to make sure all the indicators are on the same scale before combining or comparing them. We start with climate.

## Importing essential libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

### Importing CSV file for Climate data

```python
KLI_DATA = pd.read_csv('CLIMATE_FINALE.csv')
```

### Viewing Climate Dataframe's first 5 rows

```python
KLI_DATA.head()
```

### Standardizing climate columns

```python
from sklearn.preprocessing import StandardScaler

cols_to_standardize = ['ndvi', 'no2', 'co', 'so2', 'o3', 'pm10', 'pm25', 'heat_index', 'pr']

scaler = StandardScaler()

KLI_Sta = KLI_DATA.copy()
KLI_Sta[cols_to_standardize] = scaler.fit_transform(KLI_Sta[cols_to_standardize])

KLI_Sta.head()
```

### Cleans up all column names for consistency

```python
KLI_Sta.columns = KLI_Sta.columns.str.replace(' ', '_')
```

### Saves standardized climate dataframe to csv.

```python
KLI_Sta.to_csv("CLIMATE_STANDARDIZED.csv", index=False)
```
