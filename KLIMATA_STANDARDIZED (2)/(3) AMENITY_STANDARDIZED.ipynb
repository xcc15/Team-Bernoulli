## KLIMATA ILOILO DATA PREPROCESSING (STANDARDIZATION STAGE)

### We also standardized the datasets to make sure all the indicators are on the same scale before combining or comparing them. We start with amenity.

## Importing essential libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

### Importing CSV file for Amenity data

```python
AME_DATA = pd.read_csv('AMENITY_FINAL.csv')
```

### Viewing Amenity Dataframe's first 5 rows

```python
AME_DATA.head()
```

### Standardizing Amenity columns

```python
from sklearn.preprocessing import StandardScaler

cols_to_standardize = cols_to_standardize = [
    'college_nearest',
    'community_centre_nearest',
    'school_nearest',
    'shelter_nearest',
    'town_hall_nearest',
    'university_nearest'
]

scaler = StandardScaler()

AME_Sta = AME_DATA.copy()
AME_Sta[cols_to_standardize] = scaler.fit_transform(AME_Sta[cols_to_standardize])

AME_Sta.head()
```

### Saves standardized Amenity dataframe to csv.

```python
AME_Sta.to_csv("AMENITY_STANDARDIZED.csv", index=False)
```
