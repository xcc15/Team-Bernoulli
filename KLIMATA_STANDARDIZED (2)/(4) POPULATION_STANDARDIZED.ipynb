## KLIMATA ILOILO DATA PREPROCESSING (STANDARDIZATION STAGE)

### We also standardized the datasets to make sure all the indicators are on the same scale before combining or comparing them. We start with POPULATION_STANDARDIZED.

## Importing essential libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

### Importing CSV file for POPULATION STANDARDIZED data

```python
POP_DATA = pd.read_csv('POPULATION_FINAL.csv')
```

### Viewing POPULATION STANDARDIZED Dataframe's first 5 rows

```python
POP_DATA.head()
```

### Standardizing POPULATION STANDARDIZED columns

```python
from sklearn.preprocessing import StandardScaler

cols_to_standardize = [
    'pop_count_total',
    'pop_count_mean',
    'pop_count_median',
    'pop_count_stdev',
    'pop_count_min',
    'pop_count_max'
]

scaler = StandardScaler()

POP_Sta = POP_DATA.copy()
POP_Sta[cols_to_standardize] = scaler.fit_transform(POP_Sta[cols_to_standardize])

POP_Sta.head()
```

### Saves standardized POPULATION STANDARDIZED dataframe to csv.

```python
POP_Sta.to_csv("POP_STANDARDIZED.csv", index=False)
```
