## KLIMATA ILOILO DATA PREPROCESSING (MERGING)

## Importing essential libraries

```python
import pandas as pd
import numpy as np
```

### Importing CSV file for Air Quality and Location

```python
air_df = pd.read_csv("AIRQUALITY.csv")
location_df = pd.read_csv("LOCATION.csv")

print(air_df.head())
print(location_df.head())
```

### Importing CSV file for Air Quality and Location

```python
AIRQUALITY_ILOILO = pd.merge(air_df, location_df, on='adm4_pcode', how='inner')
```

### Viewing Air Qulaity Dataframe's first 10 rows

```python
AIRQUALITY_ILOILO.head(10)
```

### Selecting rows that has Iloilo City as city names

```python
AIRQUALITY_ILOILO = AIRQUALITY_ILOILO[AIRQUALITY_ILOILO['adm3_en'] == 'iloilo city']
```

### Dropping irrelevant columns

```python
AIRQUALITY_ILOILO = AIRQUALITY_ILOILO.drop(columns=['unnamed:_0', 'adm1_en', 'adm1_pcode', 'adm2_en', 'adm2_pcode', 'adm3_en', 'adm3_pcode'])
```

### Viewing updated Air Qulaity Dataframe's first 5 rows

```python
AIRQUALITY_ILOILO.head()
```

### Saves Air Quality dataframe to csv.

```python
AIRQUALITY_ILOILO.to_csv('AQ_ILOILO.csv', index=False)
```
