# KLIMATA ILOILO - DATA PREPROCESSING (MERGING OF AIR QUALITY, HEAT, RAIN DATA)

```python
import pandas as pd
import numpy as np
```

```python
air_df = pd.read_csv('AQ_ILOILO.csv')
atmos_df = pd.read_csv('HEAT_INDEX_RAIN.csv')
```

```python
air_df.head(10)
```

```python
atmos_df.head(10)
```

```python
print(air_df.shape)
print(atmos_df.shape)
```

```python
merged_df = pd.merge(
    air_df,
    atmos_df,
    on=['adm4_pcode', 'date'],
    how='inner'
)
```

```python
merged_df.shape
```

```python
merged_df.head(10)
```

```python
merged_df = merged_df.drop(columns=['adm4_en_x', 'brgy_total_area_x', 'uuid_y' ])
```

```python
merged_df = merged_df.rename(columns={'uuid_x': 'uuid', 'adm4_en_y': 'adm4_en', 'brgy_total_area_y': 'brgy_total_area'})
```

```python
merged_df.to_csv("MASTER_DATASET_CLIMATE.csv", index=False)
```
